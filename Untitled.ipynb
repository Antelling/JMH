```python
import json
import numpy as np
```

```python
scores = []
times = []
for file in [1, 2, 3, 4, 5, 6, 8, 9]:
    segments = []
    file = np.array(
        json.loads(open("results/dataset_timed_repair_" + str(file) + ".json", "r").read())["triplicate_prob_repair"])
    for problem in range(6):
        segments.append(file[list(range(problem, 90, 6))])
    scores.append([[p[0] for p in d] for d in segments])
    times.append([[p[1] for p in d] for d in segments])
```

```python
print(json.dumps([np.array(scores).tolist(), np.array(times).tolist()]))
```
